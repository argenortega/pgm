```python
%pylab inline
%matplotlib inline

import numpy as np
import networkx as nx
```

```python
nodes = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'd1', 'd2', 'd3', 'd4', 'v1', 'v2', 'v3', 'v4']
edges = [('a', 'c'), ('b', 'c'), ('b', 'd1'), ('b', 'd'), ('d1', 'v1'), ('d1', 'd'),
         ('c', 'e'), ('d', 'e'), ('e', 'g'), ('e', 'd2'), ('f', 'd2'), ('f', 'h'), 
         ('g', 'd4'), ('g', 'i'), ('d2', 'i'), ('d2', 'd3'), ('h', 'k'), ('h', 'j'),
         ('d3', 'k'), ('d3', 'v2'), ('d3', 'd4'), ('d4', 'l'), ('i', 'l'), ('j', 'v3'),
         ('k', 'v3'), ('l', 'v4')]

net = nx.MultiDiGraph()
net.add_nodes_from(nodes)
net.add_edges_from(edges)

nx.draw(net)
```

```python
moral = net.to_undirected()
nx.draw(moral)
```

```python
nodes = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'd1', 'd2', 'd3', 'd4', 'v1', 'v2', 'v3', 'v4']
chance =['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l']
decision = ['d1', 'd2', 'd3', 'd4']
utility = ['v1', 'v2', 'v3', 'v4']

edges = [('a', 'c'), ('b', 'c'), ('b', 'd1'), ('b', 'd'), ('d1', 'v1'), ('d1', 'd'),
         ('c', 'e'), ('d', 'e'), ('d', 'f'), ('e', 'g'), ('e', 'd2'), ('f', 'd2'), 
         ('f', 'h'), ('g', 'd4'), ('g', 'i'), ('d2', 'i'), ('d2', 'd3'), ('h', 'k'), 
         ('h', 'j'), ('d3', 'k'), ('d3', 'v2'), ('d3', 'd4'), ('d4', 'l'), ('i', 'l'), 
         ('j', 'v3'), ('k', 'v3'), ('l', 'v4')]

net = nx.DiGraph()
net.add_nodes_from(nodes)
net.add_edges_from(edges)

pos = nx.graphviz_layout(net, prog='dot')
nx.draw_networkx_edges(net, pos=pos)
nx.draw_networkx_labels(net, pos=pos)
nx.draw_networkx_nodes(net, pos, nodelist=chance, node_size=300)
nx.draw_networkx_nodes(net, pos, nodelist=decision, node_size=300, node_shape='s')
nx.draw_networkx_nodes(net, pos, nodelist=utility, node_size=300, node_shape='D')
```

```python
moral = net.to_undirected()
nx.draw(moral)
```
